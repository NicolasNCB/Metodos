#2.0.1 Números Factoriales
1. Escriba una función que calcule el factorial de n, con n ∈ N.




```python
def factorial (n:int)-> int:
  if n==0:
    fact=1
  else:
    fact=n
    i=1
    while (n-i)>0:
      fact=fact*(n-i)
      i+=1
  return fact

```

Encuentre los veinte primeros factoriales

```python

veinte = np.arange(1, 21)
factoriales=[]
for numero in veinte:
  factoriales.append(factorial(numero))

print(np.array(factoriales))
```

#2.0.3 Sucesión de Fibonacci
1. Encuentre los 20 primeros terminos de esta sucesión.


```python
def Fibonacci (n:int)->int:
  serie=[0,1]
  while len(serie)<n:
    serie.append(serie[len(serie)-1]+serie[len(serie)-2])
  return serie
print(np.array(Fibonacci(20)))
```

Grafique la sucesión de fibonacci

```python
fig, (punto, continua) = plt.subplots(2)
fig.suptitle('Gráfica de la Sucesión de Fibonacci')
punto.scatter(np.arange(0,20), Fibonacci(20))
continua.plot(Fibonacci(20))
```

```python
class Particle:
    
    def __init__(self,r0,v0,a0,t,m=1,radius=1,Id=0):
        
        self.dt = t[1] - t[0]
        
        self.r = r0
        self.v = v0
        self.a = a0
        
        self.R = np.zeros((len(t),len(r0)))
        self.V = np.zeros_like(self.R)
        self.A = np.zeros_like(self.R)
        
        self.radius = radius
        
    def Evolution(self,i):
        
        self.SetPosition(i)
        self.SetVelocity(i)
        
        self.r += self.dt*self.v
        self.v += self.dt*self.a
        
    def SetPosition(self,i):
        self.R[i] = self.r
        
    def GetPosition(self,scale=1):
        return self.R[::scale]
    
    def SetVelocity(self,i):
        self.V[i] = self.v

    def GetVelocity(self,scale=1):
        return self.V[::scale]

    def CheckLimits(self,Limits):
        
        for i in range(2):
        
            if self.r[i] + self.radius > Limits[i][1] and self.v[i] > 0.:
                self.v[i] = -1.0*self.v[i]
            if self.r[i] - self.radius < Limits[i][0] and self.v[i] < 0.:
                self.v[i] = -1.0*self.v[i]
```
